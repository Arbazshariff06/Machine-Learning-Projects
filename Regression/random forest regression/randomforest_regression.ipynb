# ðŸŒ¾ Crop Yield Prediction using Random Forest Regressor


## ---- Import Necessary Libraries ----


```python
import pandas as pd  # For data manipulation and analysis
import numpy as np   # For numerical computations
import matplotlib.pyplot as plt  # For visualization
import seaborn as sns  # For advanced plots
```


## ---- Load the Dataset ----

```python
dataset = pd.read_csv('crop_yield.csv')
dataset.head()
```

```python
# Display first few rows
dataset.info()
```

```python
# Show statistical summary of numerical columns
dataset.describe()
```

```python
# Check for missing values in each column
dataset.isnull().sum()
```

# ---- Encode Categorical Variables ----


```python
from sklearn.preprocessing import LabelEncoder
# Encode categorical columns into numeric values
lb = LabelEncoder()
for col in ["Region", "Soil_Type","Crop", "Weather_Condition"]:
    dataset[col] = lb.fit_transform(dataset[col])
```

## ---- Visualization Section ----


```python
# --- Distribution of Target Variable ---
plt.figure(figsize=(8,5))
sns.histplot(dataset['Yield_tons_per_hectare'], kde=True, bins=30)
plt.title("Distribution of Crop Yield")
plt.xlabel("Yield (tons/ha)")
plt.ylabel("Frequency")
plt.show()
```

```python
# --- Correlation Analysis (Numerical Features) ---
numeric_cols = ['Rainfall_mm', 'Temperature_Celsius', 'Days_to_Harvest', 'Yield_tons_per_hectare']
plt.figure(figsize=(8,6))
sns.heatmap(dataset[numeric_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# --- Countplots for Categorical Features ---
categorical_cols = ['Region', 'Soil_Type', 'Crop', 'Weather_Condition', 'Fertilizer_Used', 'Irrigation_Used']
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=dataset, x=col)
    plt.title(f"Count of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# --- Boxplots to Check Impact on Yield ---
for col in categorical_cols:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=col, y='Yield_tons_per_hectare', data=dataset)
    plt.title(f"Yield by {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# --- Pairplot for Numerical Features ---
sns.pairplot(dataset[numeric_cols])
plt.show()
```

```python
# --- Identify Outliers ---
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=dataset[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# ---- Correlation with Target Variable ----
dataset.corr()["Yield_tons_per_hectare"]
```

# ---- Prepare Data for Modeling ----


```python
X = dataset.drop(["Yield_tons_per_hectare"], axis = 1)
y = dataset["Yield_tons_per_hectare"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

## ---- Split Data into Training and Testing Sets ----


```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 42)
regressor.fit(X_train,y_train)
```

```python
# ---- Predict on Test Data ----
y_pred = regressor.predict(X_test)
y_pred
```

```python
# Reset index of X_test and y_test to align properly
X_test_reset = X_test.reset_index(drop=True)
y_test_reset = y_test.reset_index(drop=True)

# Create a comparison DataFrame with all features + actual + predicted
comparison_all = X_test_reset.copy()  # Start with features
comparison_all['Actual_Yield'] = y_test_reset  # Add actual target
comparison_all['Predicted_Yield'] = y_pred     # Add predicted target

# Display the first 20 rows
print(comparison_all.head(20))
```

```python
# Simple comparison DataFrame (Actual vs Predicted only)
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
comparison
```

```python
# Calculate evaluation metrics
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"RÂ² Score: {r2:.3f}")
print(f"Mean Absolute Error: {mae:.3f}")
print(f"Mean Squared Error: {mse:.3f}")
```

```python
# --- Plot Actual vs Predicted ---
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.7, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # perfect prediction line
plt.xlabel("Actual Yield (tons/ha)")
plt.ylabel("Predicted Yield (tons/ha)")
plt.title("Random Forest Regression: Actual vs Predicted Yield")
plt.show()
```
